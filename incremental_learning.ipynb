```python
df.drop(columns=['Unnamed: 0.2','Unnamed: 0'])
```

```python
from sklearn.utils import all_estimators
import inspect

has_n_jobs = []
for est in all_estimators():
    s = inspect.signature(est[1])
    if 'n_jobs' in s.parameters:
        has_n_jobs.append(est)
for j in has_n_jobs:
    print(j)
```

```python
from dask_ml.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,3:], df.iloc[:,2])
```

```python
classes=y_train.unique().compute()
```

```python
import numpy
```

```python
classes=[0,1]
```

```python
from sklearn.neural_network import MLPClassifier
MLP=MLPClassifier()
```

```python
from sklearn.linear_model import SGDClassifier

est = SGDClassifier(loss='log', penalty='l2', tol=1e-3)
```

```python
from dask_ml.wrappers import Incremental
inc=Incremental(est,scoring='accuracy')
```

```python
import dask
X_train, X_test, y_train, y_test = dask.persist(X_train, X_test, y_train, y_test)
```

```python
inc.fit(X_train,y_train,classes=classes)
```

```python
inc.score(X_test, y_test)
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
clf = PassiveAggressiveClassifier(max_iter=1000, random_state=0,tol=1e-3)
inc=Incremental(clf,scoring='accuracy')
inc.fit(X_train,y_train,classes=classes)
```

```python
inc.score(X_test, y_test)
```
